将tensorflow格式转为教材中使用的network格式的输入文件

```python
# TensorFlow and tf.keras
import pickle
import gzip

import tensorflow as tf
tf.__version__
```

```python
fashion_mnist = tf.keras.datasets.fashion_mnist

(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
train_images=train_images/255
test_images=test_images/255
```

```python
train_images.shape,train_images[0],train_labels.shape,train_labels[0]
```

```python
training_data=(train_images[:50000],train_labels[:50000])
training_data[0].shape,training_data[0][0],training_data[1].shape,training_data[1][0]
```

```python
validation_data=(train_images[50000:],train_labels[50000:])
validation_data[0].shape,validation_data[0][0],validation_data[1].shape,validation_data[1][0]
```

```python
test_data=(test_images, test_labels)
```

```python
"""Dump the Fashion data as a tuple containing the training data,
the validation data, and the test data.
The ``training_data`` is a tuple with two entries.
The first entry contains the actual training images.  This is a
numpy ndarray with 50,000 entries.  Each entry is, in turn, a
numpy ndarray with 784 values, representing the 28 * 28 = 784
pixels in a single FASHION image.
The second entry in the ``training_data`` tuple is a numpy ndarray
containing 50,000 entries.  Those entries are just the digit
values (0...9) for the corresponding images contained in the first
entry of the tuple.
The ``validation_data`` and ``test_data`` are similar, except
each contains only 10,000 images.
"""
f = gzip.open('fashion.pkl.gz', 'wb')
pickle.dump((training_data, validation_data, test_data), f)
f.close()
```
